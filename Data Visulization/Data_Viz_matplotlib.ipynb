```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv(r"C:\Users\King alagbe\Downloads\adeleke.csv")
data.head()
```

```python
plt.plot(data["SAMPLE A"], label = "Sample A" , color = 'green')
plt.plot(data["SAMPLE B"], label = "Sample B"  , color = "red", linewidth = 4)
plt.plot(data["SAMPLE C"], label = "Sample C", color = "purple", linestyle = "dashed")
plt.plot(data["SAMPLE D"], label = "Sample D", color = "blue", linestyle = "dashdot")
plt.plot(data["SAMPLE E"], label = "Sample E", color = "cyan", linestyle = "dotted")
plt.plot(data["INDUSTRIAL GRADE"], label = "Industrial Grade", color = "magenta", linestyle = "solid")

plt.legend()

plt.ylabel("SAMPLES ", fontsize = 10)
plt.xlabel("PHYSIO-CHEMICAL PROPERTIES", fontsize = 10)

plt.show()
```
